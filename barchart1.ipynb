```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
 'Category':  ['60', '18 - 29', '30 - 44', '45 - 59'],
'Value1':  [71, 28,	95,	64],
'Value2':  [61,	11,	68,	36],
'Value3':  [68,	14,	71,	51],
'Value4':  [73,	20,	128, 57]
}


df = pd.DataFrame(data)


df['Cumulative_Value1'] = df['Value1'].cumsum()
df['Cumulative_Value2'] = df['Value2'].cumsum()
df['Cumulative_Value3'] = df['Value3'].cumsum()
df['Cumulative_Value4'] = df['Value4'].cumsum()


plt.barh(df['Category'], df['Cumulative_Value1'], label='Somewhat likely', alpha=0.6)
plt.barh(df['Category'], df['Value2'], left=df['Cumulative_Value1'], label='Somewhat unlikely', alpha=0.6)
plt.barh(df['Category'], df['Value3'], left=df['Cumulative_Value1'] + df['Value2'], label='Very likely', alpha=0.6)
plt.barh(df['Category'], df['Value4'], left=df['Cumulative_Value1'] + df['Value2'] + df['Value3'], label='Very unlikely', alpha=0.6)






plt.legend()
plt.title('Cumulative Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks([])

plt.show()
```
